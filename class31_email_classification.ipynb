# Spam Email Classifier

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('email.csv')
```

```python
data
```

```python
# convert ham--> 1 , spam --0
data.loc[:,'class']=data.loc[:,'Category'].map({'ham':1,'spam':0})
```

```python
data = data.dropna()
```

**Basic Data Cleaning**

```python
import re
import string
```

```python
def clean_text (text):
  text= text.lower()
  # remove html tags
  text = re.sub(r'<.*?>',"",text)
  # remove urls
  text = re.sub(r'http\S+|www\S+','',text)
  # remove emojis
  text = re.sub(r'[^x00-\x7F]+',' ',text)
  # special characters
  text = re.sub(r'[^a-z0-9\s]','',text)
  # white space
  text = re.sub(r'\s+',' ',text).strip()

  return text
```

```python
data.loc[:,'cleaned_text']=data['Message'].apply(clean_text)
```

```python
data
```

**Split data for x and y**

```python
x = data['cleaned_text']
y= data['class']
```

**Split data for training and Testing**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

# Count Vectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
x_new_train = cv.fit_transform(x_train)
x_new_test = cv.transform(x_test)
```

```python
# fit_transform : Learn + transform
# transform : It just transforms
```

**Naive Bayes**

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(x_new_train,y_train)
```

```python
y_pred = model.predict(x_new_test)
```

```python
# confusion matrix
from sklearn.metrics import confusion_matrix,accuracy_score,recall_score,precision_score,f1_score
confusion_matrix(y_test,y_pred)
```

```python
# accuracy
accuracy_score(y_test,y_pred)
```

```python
# recall
recall_score(y_test,y_pred)
```

```python
_# precision
precision_score(y_test,y_pred)
```

```python
# f1
f1_score(y_test,y_pred)
```

**Predictions**

```python
msg = input("enter the message")
text = cv.transform([msg])
pred = model.predict(text)
if pred[0]==0:
  print("Spam")
else:
  print("Ham")
```

```python
msg = input("enter the message")
text = cv.transform([msg])
pred = model.predict(text)
if pred[0]==0:
  print("Spam")
else:
  print("Ham")
```

# TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf = TfidfVectorizer()
x_train_tf = tf.fit_transform(x_train)
x_test_tf = tf.transform(x_test)
```

```python
model = MultinomialNB()
model.fit(x_train_tf,y_train)
y_pred = model.predict(x_test_tf)
```
