```python
# 1. Initializing a list
my_list = [10, 40, 20, 30, 40, 50]
print("Initial list:", my_list)

# 2. append(): Adds an element to the end of the list
my_list.append(60)
print("After append(60):", my_list)

# 3. extend(): Appends elements from another iterable to the end of the list
new_elements = [70, 80]
my_list.extend(new_elements)
print("After extend([70, 80]):", my_list)

# 4. insert(): Inserts an element at a specified index
my_list.insert(1, 15) # Insert 15 at index 1
print("After insert(1, 15):", my_list)

# 5. remove(): Removes the first occurrence of a specified value
my_list.remove(40) # Removes the first 40
print("After remove(40):", my_list)

# 6. pop(): Removes and returns the element at a specified index (or the last element if no index is given)
popped_item = my_list.pop() # Removes and returns the last element
print("After pop() (popped item:", popped_item, "):", my_list)
popped_item_at_index = my_list.pop(0) # Removes and returns the element at index 0
print("After pop(0) (popped item:", popped_item_at_index, "):", my_list)

# 7. clear(): Removes all items from the list
list_to_clear = [1, 2, 3]
list_to_clear.clear()
print("After clear() on a new list:", list_to_clear)

# Re-initialize list for further demonstrations
my_list = [10, 40, 20, 30, 40, 50, 40]
print("Re-initialized list:", my_list)

# 8. index(): Returns the index of the first occurrence of a specified value
index_of_20 = my_list.index(20)
print("Index of 20:", index_of_20)
try:
    # Can specify start and end index for search
    index_of_40_after_index_2 = my_list.index(40, 3) # Search for 40 starting from index 3
    print("Index of 40 after index 2:", index_of_40_after_index_2)
except ValueError as e:
    print(f"Error finding 40 after index 2: {e}")

# 9. count(): Returns the number of times a specified value appears in the list
count_of_40 = my_list.count(40)
print("Count of 40:", count_of_40)

# 10. sort(): Sorts the list in ascending order by default (in-place)
my_list.sort()
print("After sort():", my_list)

# Sort in descending order
my_list.sort(reverse=True)
print("After sort(reverse=True):", my_list)

# 11. reverse(): Reverses the order of elements in the list (in-place)
my_list.reverse()
print("After reverse():", my_list)

# 12. copy(): Returns a shallow copy of the list
original_list = [1, 2, 3]
copied_list = original_list.copy()
print("Original list:", original_list)
print("Copied list:", copied_list)

# Demonstrate that it's a shallow copy (modifying copy doesn't affect original for primitives)
copied_list.append(4)
print("Original list after modifying copy:", original_list)
print("Copied list after modifying copy:", copied_list)

# For nested lists, it's a shallow copy
nested_original = [[1], [2]]
nested_copy = nested_original.copy()
nested_copy[0].append(10)
print("Nested original after modifying nested copy:", nested_original)
print("Nested copy after modifying nested copy:", nested_copy)

# len(): (Built-in function, not a method, but commonly used)
list_length = len(my_list)
print("Length of the final list:", list_length)
```
