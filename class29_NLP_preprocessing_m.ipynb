# Basic Data Cleaning

```python
text = "        <p>Hello!!! 😊 Visit https://example.com</p> NLP is FUN!!! #AI @(2026),..   "
```

```python
def clean_text (text):
  text= text.lower()
  # remove html tags
  text = re.sub(r'<.*?>',"",text)
  # remove urls
  text = re.sub(r'http\S+|www\S+','',text)
  # remove emojis
  text = re.sub(r'[^x00-\x7F]+',' ',text)
  # special characters
  text = re.sub(r'[^a-z0-9\s]','',text)
  # white space
  text = re.sub(r'\s+',' ',text).strip()

  return text
```

```python
clean_text(text)
```

# POS Tagging

```python
from nltk import pos_tag
text = ' NLP is easy and very interesting to explore more about it'
words = word_tokenize(text)
pos_tag(words)
```

| Word        | Tag | Meaning                       |
| ----------- | --- | ----------------------------- |
| NLP         | NNP | Proper Noun (name of a field) |
| is          | VBZ | Verb (3rd person singular)    |
| very        | RB  | Adverb                        |
| interesting | JJ  | Adjective                     |
| and         | CC  | Conjunction                   |
| easy        | JJ  | Adjective                     |
| to          | TO  | "to" (infinitive marker)      |

# Named Entity Recognition

```python
from nltk import ne_chunk
text = "John is working at Google company as HR in Bangalore"
words = word_tokenize(text)
tag = pos_tag(words)
ner = ne_chunk(tag)
print(ner)
```

```python
import spacy
model = spacy.load('en_core_web_sm')
```

**Word Tokenization**

```python
text = 'NLP is easy'
words = model(text)
[x for x in words]
```

**Sentence Tokenization**

```python
text = 'NLP is easy.Python is hard'
sentences = model(text)
[x.text for x in sentences.sents]
```

**Stop words**

```python
text = 'NLP is very easy and interesting'
words = model(text)
[x.text for x in words if not x.is_stop]
```

**lemmetization**

```python
text = 'playing plays played studying studies good better'
words = model(text)
[x.lemma_ for x in words]
```

**Pos Tagging**

```python
text = 'I have started studying NLP and its very intresting'
words = model(text)
{x:x.pos_ for x in words if not x.is_stop}
```

**NER**

```python
text = "John is working at Google company as HR in Bengaluru"
words = model(text)
{x.text:x.label_ for x in words.ents}
```
